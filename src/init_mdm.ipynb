```python
def get_data(df, column, values):
    if isinstance(values, str):
        values = [values]
    return df[df[column].isin(values)]
```

вычисление top_primary - предварительно вычисленная дата окончания, там где явно задано окончание, то  его указываем, там где не указано, то по договору лизинга 
Вынесла отдельно, как часто повторяющийся код

```python
def determine_stop_primary(df):
    """stop_primary - предварительно вычисленная дата окончания, там где явно задано окончание, то  его указываем, там где не указано, то по договору лизинга"""
    mask = (df['cession_date'].notnull())
    df.loc[mask, 'stop_primary'] = df.loc[mask, 'cession_date']

    mask = (df['new_terminationleasing_date'].notnull())
    df.loc[mask, 'stop_primary'] = df.loc[mask, 'new_terminationleasing_date']

    mask = (df['stop_primary'].isnull())
    df.loc[mask, 'stop_primary'] = df.loc[mask, 'new_leasing_enddate']

    mask = (df['stop_primary'].isnull())
    mask_stop = (df['new_kreditmessage_type'] == 'StopFinancialLeaseContract')
    df.loc[mask & mask_stop, 'stop_primary'] = df.loc[mask & mask_stop, 'new_kreditmessage_date']
```

Меняем на числовые значения в днях timedelta64 в фичах

```python
def days_substitute(df_features):
    for delta_col in df_features.columns:
        if (is_timedelta64_dtype(df_features[delta_col])):
            df_features[delta_col] = df_features[delta_col].dt.days
            df_features[delta_col] = df_features[delta_col].fillna(0)
    return df_features
```
